#Traitement des valeurs extremes

Pour traiter ces valeurs, nous pouvons utiliser les methodes suivantes.

#1)Bases sur les quartiles


```python
print(data['Tea'].quantile(0.10))
print(data['Tea'].quantile(0.90))
```

```python
import numpy as np
data["Tea"] = np.where(data["Tea"] <54.0, 54.0,data['Tea'])
data["Tea"] = np.where(data["Tea"] >98.5, 98.5,data['Tea'])
print(data['Tea'].skew())
```

#2) Transformation logarithmique

```python
data["Log_Biscuits"] = data["Biscuits"].map(lambda i: np.log(i) if i > 0 else 0) 
print(data['Biscuits'].skew())
print(data['Log_Biscuits'].skew())
```

#3) Enlever les valeurs extremes

```python
index = data[(data['Tea'] >= 80)|(data['Tea'] <= 18)].index
data.drop(index, inplace=True)
data['Tea'].describe()
```
