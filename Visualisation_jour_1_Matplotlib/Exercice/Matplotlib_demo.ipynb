#Definition de la visualisation.
La visualisation des données est la discipline qui consiste à essayer de comprendre les données en les plaçant dans un contexte visuel afin de mettre en évidence des schémas, des tendances et des corrélations qui n'auraient pas pu être détectés autrement.  La visuallisation c'est tres interessante. En effet en raison de la manière dont le cerveau humain traite l'information, il est plus facile d'utiliser des diagrammes ou des graphiques pour visualiser de grandes quantités de données complexes que de passer en revue des feuilles de calcul ou des rapports. Cette visualisation permet d'identifier les domaines qui nécessitent une attention ou une amélioration,
clarifier les facteurs qui influencent le comportement des clients. Bref
La visualisation permet de faire une exploration tres creative pour prendre des décisions biens precises et developper notre business. Comme outil nous pouvons avoir matplotlib et seaborn.

#Créer un titre et des étiquettes pour un graphe
Vous pouvez utiliser le paramètre fontdict dans xlabel(), ylabel(), et title() pour définir les propriétés de la police pour le titre et les étiquettes.

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

#Plusieurs courbes dans une figure

La fonction subplots() prend trois arguments qui décrivent la disposition de la figure.

La disposition est organisée en lignes et en colonnes, qui sont représentées par le premier et le deuxième argument.

Le troisième argument représente l'index du graphique actuel.

```python
#plot 1:
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C = np.cos(X)

plt.subplot(1, 2, 1)  #the figure has 1 row, 2 columns, and this plot is the first plot.
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-")

#plot 2:
S = np.sin(X)

plt.subplot(1, 2, 2) #the figure has 1 row, 2 columns, and this plot is the second plot.
plt.plot(X, S, color="red",  linewidth=2.5, linestyle="--")

plt.show()
```

```python
#plot
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C = np.cos(X)

plt.subplot(2, 1, 1)  #the figure has 2 row, 2 columns, and this plot is the first plot.
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-")

#plot 2:
S = np.sin(X)

plt.subplot(2, 1, 2) #the figure has 2 row, 1 columns, and this plot is the second plot.
plt.plot(X, S, color="red",  linewidth=2.5, linestyle="--")

plt.show()
```

#Dimension 3

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
ax = plt.axes(projection='3d')
z = np.linspace(0, 1, 100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)
c = x + y
ax.scatter(x, y, z, c=c) #ax.plot3D(x, y, z, 'gray') # we can change it from RGB  to gray by using 'gray'
ax.set_title('3D line plot')
plt.show()
```

#stackplot
Si vous voulez observer la composition de certaines données et leur évolution relative dans le temps. Vous pouvez utiliser la fonction plt.stackplot.


```python
 #List of 7-days
days = [x for x in range(0, 7)]
  
# List of Suspected cases
Suspected = [12, 18, 35, 50, 72, 90, 100]
  
# List of Cured Cases
Cured = [4, 8, 15, 22, 41, 55, 62]
  
# List of Number of deaths
Deaths = [1, 3, 5, 7, 9, 11, 13]
  
# Plot x-labels, y-label and data
plt.plot([], [], color ='blue', 
         label ='Suspected')
plt.plot([], [], color ='orange',
         label ='Cured')
plt.plot([], [], color ='brown',
         label ='Deaths')
  
# Implementing stackplot on data
plt.stackplot(days, Suspected, Cured, 
              Deaths, 
              colors =['blue', 'orange', 
                       'brown'])
  
plt.legend()
  
plt.title('No of Cases')
plt.xlabel('Day of the week')
plt.ylabel('Overall cases')
  
plt.show()
```
