#*Exercise 1:* 
Read Total profit of all months and show it using a line plot

Total profit data provided for each month. Generated line plot must include the following properties: –

* X label name = Month Number
* Y label name = Total profit
* The line plot graph should look like this.

![](https://pynative.com/wp-content/uploads/2019/01/matplotlib_and_pandas_exercise_1_plot_company_profit.png)

```
X=data['month_number']
Y=data['total_profit']
a = {'name':'Month Number','color':'blue','size':20}
b = {'name':'Month Number','color':'k','size':20}
plt.figure(figsize=(15,6), dpi=80)
plt.plot(X,Y, color='b',linewidth=2)
plt.xticks(X)
plt.xlabel("Month Number",fontdict=a)
plt.ylabel("Total profit",fontdict=b)
plt.title("Compagny profit by month",size=20)
```

# *Exercise Question 2:* 
Get Total profit of all months and show line plot with the following Style properties

* X label name = Month Number
* Y label name = total profit

# *Exercise Question 3: Read all product sales data and show it  using a multiline pl*

Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).

The graph should look like this.

![](https://pynative.com/wp-content/uploads/2019/01/matplotlib_and_pandas_exercise_3_multiline_plots.png)

```
X=data['month_number']
Y=data[['facecream','facewash',	'toothpaste',	'bathingsoap',	'shampoo',	'moisturizer']]
a = {'name':'Month Number','color':'blue','size':20}
b = {'name':'Month Number','color':'darkred','size':15}
plt.figure(figsize=(10,6), dpi=80)

#labels = ['facecream','facewash',	'toothpaste',	'bathingsoap',	'shampoo',	'moisturizer']
#for  y, label in zip(Y.values, labels):
plt.plot(X,Y['facecream'], marker='o',ms=15,linewidth=3 , label='facecream')
plt.plot(X,Y['facewash'], marker='o',ms=15,linewidth=3 , label='facewash')
plt.plot(X,Y['toothpaste'], marker='o',ms=15,linewidth=3 , label='toothpaste')
plt.plot(X,Y['bathingsoap'], marker='o',ms=15,linewidth=3 , label='bathingsoap')
plt.plot(X,Y['shampoo'], marker='o',ms=15,linewidth=3 , label='shampoo')
plt.plot(X,Y['moisturizer'], marker='o',ms=15,linewidth=3 , label='moisturizer')


plt.legend(loc='upper left')
plt.xlabel("Month Number",fontdict=a)
plt.ylabel("Sale unit in number",fontdict=b)

plt.title("Sale data",size=20)
plt.legend(loc='upper left')
```

#*Exercise Question 4: Read toothpaste sales data of each month and show it using a scatter plot*

Also, add a grid in the plot. gridline style should “–“.

The scatter plot should look like this.

![](https://pynative.com/wp-content/uploads/2019/01/matplotlib_and_pandas_exercise_4_show_scatter_plot.png)

```
X=data['month_number']
Y=data['toothpaste']

a = {'name':'Month Number','color':'blue','size':20}
b = {'name':'Month Number','color':'darkred','size':15}
plt.figure(figsize=(10,6), dpi=80)
plt.scatter(X,Y, label='Tooth paste sales data')
plt.grid()

plt.legend(loc='upper left')
plt.xlabel("Month Number",fontdict=a)
plt.ylabel("Number of unit sold",fontdict=b)

plt.title("Tooth paste sales data",size=20)
plt.legend(loc='upper left')
```

#*Exercise Question 5: Read face cream and facewash product sales data and show it using the bar chart*

The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

The bar chart should look like this.

![](https://pynative.com/wp-content/uploads/2019/01/matplotlib_panadas_exercise_5_show_bar_chart.png)

```
X=data['month_number']
Y=data['facecream']
Z=data['facewash']

a = {'name':'Month Number','color':'blue','size':20}
b = {'name':'Month Number','color':'darkred','size':15}
plt.figure(figsize=(10,6), dpi=80)
plt.bar(X -  0.125,Y, width=0.4,label= 'Face cream data')
plt.bar(X +  0.125,Z, width=0.4, label= 'Face wash data')
plt.grid()

plt.legend(loc='upper left')
plt.xlabel("Month Number",fontdict=a)
plt.ylabel("Sale unit in number",fontdict=b)

plt.title("Face cream and face wash sales data",size=20)
plt.legend(loc='upper left')
plt.show()
```
