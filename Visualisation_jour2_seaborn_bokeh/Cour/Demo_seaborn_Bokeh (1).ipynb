#Seaborn

```python
# loading dataset 
data = sb.load_dataset("iris") 
data.head()
data["species"].unique()
data["species"].value_counts()
```

#Bar plot

```python
sb.barplot(x='species', y='sepal_length', data=data)
plt.show()
plt.savefig('g')
```

#Hisgramme

```python
sb.histplot(x='species', y='sepal_width', data=data)
plt.show()
```

#Countplot
C'est une representation qui fait le compte des elements d'une variable categorique

```python
sb.countplot(x='species', data=data)
plt.show()
```

#Boxplot(describe distribution of data)

```python
sb.boxplot(x='species', y='sepal_width', data=data)
#plt.show()
```

#nuage de points(scatter)
cela peut aider a determiner une relation entre deux variables. Chaque point represente une observation.

```python
sb.scatterplot(x='sepal_length', y='sepal_width', data=data)
plt.show()
```

#nuage de point basee sur les variables gategory(stripplot)

```python
sb.stripplot(x='species', y='sepal_width', data=data)
plt.show()
```

#displot 
reprentation de la densite

```python
sb.distplot(data['sepal_width'])
#data[['sepal_width',""]]
plt.show()
```

#Pairplot
cela permet de detecter facilement les relations deux a deux au niveau de notre jeu de donnees.

```python
sb.pairplot(data=data, hue='species')
plt.show()
```

# draw lineplot 

```python
sb.lineplot(x="sepal_length", y="sepal_width", data=data)
# setting the title using Matplotlib
plt.title('Titre de la courbe')
  
plt.show()
```

```python
sb.lineplot(x="sepal_length", y="sepal_width", data=data,marker='o') 
  
# setting the x limit of the plot
plt.xlim(5)
plt.ylim(3)
  
plt.show()
```

#Bokeh
https://docs.bokeh.org/en/latest/docs/first_steps.html#first-steps

```python
from bokeh.plotting import figure, output_file,show
output_file("line2.html")

# prepare some data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# create a new plot with a title and axis labels
p = figure(title="Simple line example", x_axis_label="x", y_axis_label="y")

# add a line renderer with legend and line thickness
p.line(x, y, legend_label="Temp.", line_width=2)

# show the results
show(p)
```

```python
from bokeh.plotting import figure, output_file, show

# output to static HTML file
output_file("line3.html")

p = figure(plot_width=400, plot_height=400)

# add a circle renderer with a size, color, and alpha
p.circle([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], size=20, color="navy", alpha=0.5)

# show the results
show(p)
```

```python
from bokeh.plotting import figure, output_file, show

output_file("line.html")

p = figure(plot_width=400, plot_height=400)

# add a line renderer
p.line([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], line_width=2)

show(p)
```

```python
from bokeh.plotting import figure, output_file, show

output_file("patch.html")

p = figure(plot_width=400, plot_height=400)

p.multi_line([[1, 3, 2], [3, 4, 6, 6]], [[2, 1, 4], [4, 7, 8, 5]],
             color=["firebrick", "navy"], alpha=[0.8, 0.3], line_width=4)
show(p)
```

```python
from bokeh.plotting import figure, output_file, show

output_file('vbar.html')

p = figure(plot_width=400, plot_height=400)
p.vbar(x=[1, 2, 3], width=0.5, bottom=0,
       top=[1.2, 2.5, 3.7], color="firebrick")

show(p)
```

```python
from bokeh.plotting import figure, output_file, show

output_file('hbar.html')

p = figure(plot_width=400, plot_height=400)
p.hbar(y=[1, 2, 3], height=0.5, left=0,
       right=[1.2, 2.5, 3.7], color="navy")

show(p)
```

```python
from bokeh.plotting import figure
from bokeh.io import output_notebook, push_notebook, show

output_notebook()

plot = figure(plot_height= 250, plot_width = 300)
plot.circle([1,2,3], [4,6,5], size = 10, color ='red')

handle = show(plot, notebook_handle=True)

# Update the plot title in the earlier cell
plot.title.text = "Test Title"
push_notebook(handle = handle)
```
